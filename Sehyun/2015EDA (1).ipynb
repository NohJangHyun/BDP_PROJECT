```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
lin_reg = linear_model.LinearRegression()
```

```python
data2015
```

```python
zero_index = data2015[data2015['이용거리']==0].index
df = data2015.drop(zero_index)
df
```

```python
X = df[['이용시간']]
y = df[['이용거리']]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=1004)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lin_reg_model = lin_reg.fit(X_train,y_train)
```

```python
y_pred = lin_reg_model.predict(X_test)
y_pred
```

```python
print(r2_score(y_test,y_pred))
```

R2가 0.4782247208332421 밖에 나오지 않지만.... 울며 겨자먹기로 이 값이라도 쓴다.
